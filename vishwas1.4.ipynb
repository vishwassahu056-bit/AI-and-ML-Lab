```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame([
    {"customerID": 1, "customername": "Rahul", "product": "Bread", "qty": 2, "cost": 60, "category": "food"},
    {"customerID": 2, "customername": "Sneha", "product": "Butter", "qty": 1, "cost": 120, "category": "spread"},
    {"customerID": 3, "customername": "Aman", "product": "Juice", "qty": 3, "cost": 150, "category": "drink"},
    {"customerID": 4, "customername": "Neha", "product": "Oats", "qty": 2, "cost": 180, "category": "healthy"},
    {"customerID": 5, "customername": "Pooja", "product": "Chips", "qty": 4, "cost": 100, "category": "fastfood"},
    {"customerID": 6, "customername": "Ravi", "product": "Dryfruit", "qty": 1, "cost": 600, "category": "healthy"},
    {"customerID": 7, "customername": "Rohit", "product": "Milk", "qty": 5, "cost": 250, "category": "drink"}
])

print(df)
plt.scatter(df["product"], df["category"])
plt.xlabel("Product")
plt.ylabel("Category")
plt.show()

plt.scatter(df["qty"], df["category"])
plt.xlabel("Quantity")
plt.ylabel("Category")
plt.show()


plt.boxplot(df["cost"])
plt.ylabel("Cost")
plt.show()


mean = df["cost"].mean()
std = df["cost"].std()

df["z_score"] = (df["cost"] - mean) / std

print(df["z_score"])
```
