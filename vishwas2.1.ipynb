```python
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E'],
    'C': ['A', 'F'],
    'D': ['A', 'G'],
    'E': ['B', 'H'],
    'F': ['C', 'H'],
    'G': ['D', 'H'],
    'H': ['E', 'F', 'G']
}
def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False
dfs('A', 'H', graph)
```
