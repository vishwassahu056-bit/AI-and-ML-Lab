```python
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E'],
    'C': ['A', 'F'],
    'D': ['A', 'G'],
    'E': ['B', 'H'],
    'F': ['C', 'H'],
    'G': ['D', 'H'],
    'H': ['E', 'F', 'G']
}
def bfs(graph, start_node, goal_node):
    vis = set()
    queue = [start_node]
    path = []

    vis.add(start_node)

    while queue:
        node = queue.pop(0)
        path.append(node)

        if node == goal_node:
            return path

        for neighbour in graph[node]:
            if neighbour not in vis:
                vis.add(neighbour)
                queue.append(neighbour)

    return path
r = bfs(graph, 'A', 'H')
print(r)
```
