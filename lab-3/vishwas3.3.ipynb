```python

graph = {
    'A': ['B'],
    'B': ['A', 'C', 'D'],
    'C': ['B', 'D'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['D', 'F'],
    'F': ['D', 'E']
}

def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False

dfs('A', 'E', graph)
```
