```python

import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Type of House": [
        "PG", "Banglore", "Hostel", "Penthouse",
        "PG", "Hostel", "Banglore", "Penthouse",
        "PG", "Hostel", "Banglore", "Penthouse",
        "PG", "Hostel", "Banglore"
    ],
    "Area of House": [
        300, 1200, 450, 2500,
        400, 600, 1500, 3000,
        350, 550, 1800, 2800,
        500, 700, 2000
    ],
    "Size": [
        "1 Sharing", "2 BHK", "2 Sharing", "4 BHK",
        "2 Sharing", "3 Sharing", "3 BHK", "5 BHK",
        "1 Sharing", "2 Sharing", "4 BHK", "4 BHK",
        "3 Sharing", "3 Sharing", "3 BHK"
    ],
    "Cost": [
        7000, 4500000, 12000, 12000000,
        10000, 18000, 6000000, 15000000,
        9000, 16000, 7500000, 13500000,
        14000, 20000, 8000000
    ]
}

df = pd.DataFrame(data)

df["Normalized Area"] = (
    (df["Area of House"] - df["Area of House"].min()) /
    (df["Area of House"].max() - df["Area of House"].min())
)

df["Normalized Cost"] = (
    (df["Cost"] - df["Cost"].min()) /
    (df["Cost"].max() - df["Cost"].min())
)

print("Dataset with Normalized Values:")
print(df)
plt.figure()
plt.scatter(df["Area of House"], df["Cost"])
plt.xlabel("Area of House")
plt.ylabel("Cost")
plt.title("Area vs Cost (Original Data)")
plt.show()

plt.figure()
plt.scatter(df["Normalized Area"], df["Normalized Cost"])
plt.xlabel("Normalized Area")
plt.ylabel("Normalized Cost")
plt.title("Normalized Area vs Normalized Cost")
plt.show()
```
