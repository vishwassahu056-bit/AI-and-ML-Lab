```python

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['B', 'C', 'D', 'E']
}

def bfs(graph, start_node, goal_node):
    vis = set()
    queue = [start_node]
    path = []

    vis.add(start_node)

    while queue:
        node = queue.pop(0)
        path.append(node)

        if node == goal_node:
            return path

        for neighbor in graph[node]:
            if neighbor not in vis:
                vis.add(neighbor)
                queue.append(neighbor)

    return path
r = bfs(graph, 'A', 'F')
print(r)
```
